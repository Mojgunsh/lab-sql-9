```python
#How many distinct (different) actors' last names are there?
data = pd.read_sql_query('select count(distinct(last_name)) as count from actor', engine)
display(data)
```

```python
#Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
query = ''' SELECT *,
CASE
WHEN DATE_FORMAT(rental_date,'%%W') ='Tuesday'  then 'Workday'
WHEN DATE_FORMAT(rental_date,'%%W') = "wednesday" then 'Workday'
WHEN DATE_FORMAT(rental_date,'%%W') = "Thursday" then 'Workday'
WHEN DATE_FORMAT(rental_date,'%%W') = "Monday" then 'Workday'
WHEN DATE_FORMAT(rental_date,'%%W') = "Friday" then 'Workday'
WHEN DATE_FORMAT(rental_date,'%%W') = "Saturday" then 'weekend'
WHEN DATE_FORMAT(rental_date,'%%W') = "Sunday" then 'weekend'
ELSE 'No status'
END AS 'W_day',DATE_FORMAT(rental_date,'%%M') as month ,
DATE_FORMAT(rental_date,'%%W') as day
FROM rental;'''

data = pd.read_sql_query(query, engine)
display(data)
```

```python
#Get all films with ARMAGEDDON in the title.
query='''SELECT * FROM film where(title like '%%ARMAGEDDON%%');'''
data=pd.read_sql_query(query,engine)
display(data)
```

```python
#Get 10 the longest films.
query=''' select title,length from film
order by length Desc limit 10;'''
data=pd.read_sql_query(query,engine)
display(data)
```

```python
# How many films include Behind the Scenes content?
query='''SELECT count(*) FROM film where
(special_features like ('%%Behind the Scenes%%'));'''
data=pd.read_sql_query(query,engine)
display(data)
```

```python
#Which kind of movies (rating) have a mean duration of more than two hours?
query='''select rating ,round(avg(length),2) as Average_len
from film
group by rating
having Average_len>120;'''
data=pd.read_sql_query(query,engine)
display(data)
```

```python
#Rank films by length (filter out the rows that have nulls or 0s in length column). 
#In your output, only select the columns title, length, and the rank.
query='''select title,length,rank() over (ORDER BY length desc) rank_
from film
WHERE (title IS NOT NULL) AND (title != '  ')
order by length desc;'''
data=pd.read_sql_query(query,engine)
display(data)
```
